1. We first **clean** the dataset: clear the dataset "life_expectancy" from missing data and correct the colum names for spaces before or after the name.

```python
import pandas as pd
life_expectancy = pd.read_csv("https://raw.githubusercontent.com/NHameleers/dtz2025-datasets/master/CountryHealthFactors.csv")
life_expectancy.head()
#https://www.kaggle.com/datasets/kumarajarshi/life-expectancy-who
```

```python
life_expectancy.describe()
```

```python
life_expectancy_strip= life_expectancy.rename(columns=str.strip)
```

```python
life_expectancy.isnull().sum()
```

```python
LE_clean1 = life_expectancy_strip.dropna(thresh=1, subset=['BMI'])
```

```python
LE_clean2 = LE_clean1.dropna(thresh=1, subset=['GDP'])
```

```python
LE_cleaned = LE_clean2.dropna(thresh=1, subset=['Schooling'])
```

BMI, GDP, and Schooling now have no more missing values:

```python
LE_cleaned.isnull().sum()
```






2. Next, we create a **new variable**, called "education", which divides the years of Schooling into primary (below 8 years of schooling) and secondary (8 or more years of schooling) education. We then select only relevant columns.

```python
LE_cleaned.loc[LE_cleaned.Schooling <8, "education"] = "primary"
LE_cleaned.loc[LE_cleaned.Schooling ==8, "education"] = "primary"
LE_cleaned.loc[LE_cleaned.Schooling >8, "education"] = "secondary"

LE_cleaned.head()
```
