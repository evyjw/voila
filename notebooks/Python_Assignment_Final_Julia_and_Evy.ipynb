```python
import pandas as pd
life_expectancy = pd.read_csv("https://raw.githubusercontent.com/NHameleers/dtz2025-datasets/master/CountryHealthFactors.csv")
life_expectancy.head()
#https://www.kaggle.com/datasets/kumarajarshi/life-expectancy-who
```

```python
life_expectancy.describe()
```

```python
life_expectancy_strip= life_expectancy.rename(columns=str.strip)
```

```python
life_expectancy.isnull().sum()
```

```python
LE_clean1 = life_expectancy_strip.dropna(thresh=1, subset=['BMI'])
```

```python
LE_clean2 = LE_clean1.dropna(thresh=1, subset=['GDP'])
```

```python
LE_cleaned = LE_clean2.dropna(thresh=1, subset=['Schooling'])
```

BMI, GDP, and Schooling now have no more missing values:






2. Next, we create a **new variable**, called "education", which divides the years of Schooling into primary (below 8 years of schooling) and secondary (8 or more years of schooling) education. We then select only relevant columns.

```python
def visualize():
  print("Here you can see the relationship between GDP and BMI for almost 200 different countries, color-coded by education level. Please choose a year, or 'refresh' to choose a different year!")
  def drop_Year(change):
    Years = change.new
    if Years == "2000":
      clear_output()
      display(dropdown_menu)
      display(Chart2000)
    elif Years == "2001":
      clear_output()
      display(dropdown_menu)
      display(Chart2001)
    elif Years == "2002":
      clear_output()
      display(dropdown_menu)
      display(Chart2002)
    elif Years == "2003":
      clear_output()
      display(dropdown_menu)
      display(Chart2003)
    elif Years == "2004":
      clear_output()
      display(dropdown_menu)
      display(Chart2004)
    elif Years == "2005":
      clear_output()
      display(dropdown_menu)
      display(Chart2005)
    elif Years == "2006":
      clear_output()
      display(dropdown_menu)
      display(Chart2006)
    elif Years == "2007":
      clear_output()
      display(dropdown_menu)
      display(Chart2007)
    elif Years == "2008":
      clear_output()
      display(dropdown_menu)
      display(Chart2008)
    elif Years == "2009":
      clear_output()
      display(dropdown_menu)
      display(Chart2009)
    elif Years == "2010":
      clear_output()
      display(dropdown_menu)
      display(Chart2010)
    elif Years == "2011":
      clear_output()
      display(dropdown_menu)
      display(Chart2011)
    elif Years == "2012":
      clear_output()
      display(dropdown_menu)
      display(Chart2012)
    elif Years == "2013":
      clear_output()
      display(dropdown_menu)
      display(Chart2013)
    elif Years == "2014":
      clear_output()
      display(dropdown_menu)
      display(Chart2014)
    elif Years == "2015":
      clear_output()
      display(dropdown_menu)
      display(Chart2015)
    elif Years == "make a choice":
      clear_output()
      display(dropdown_menu)
      AppLayout(header=header_output,
          footer=footer_output)



  list_of_years = ["make a choice", '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
  dropdown_menu = widgets.Dropdown(options=list_of_years, description =  "Select Year")

  dropdown_menu.observe(drop_Year)
  display(dropdown_menu)
```
